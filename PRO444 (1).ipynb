```python
df = wd.data('crime1')
Y = df['narr86'].to_numpy()
```

```python
matrizx1 =df[["pcnv", "avgsen", "tottime", "inc86", "pcnvsq", "inc86sq"]]
matrizx2= matrizx1.to_numpy()
```

```python
X = np.insert(matrizx2, 0, 1, axis=1)
```

```python
class reg_lin:

  def __init__(self, X, Y):                                      #Aqui hacemos lo necesario para empezar con los calculos

    self.X = np.insert(matrizx2, 0, 1, axis=1)                   #Matriz variables independientes
    self.Y = df['narr86'].to_numpy()                             #Matriz variable dependiente
    self.n = X.shape[0]                                          #Numero de observaciones
    self.k = self.X.shape[1]-1                                   #k
    self.Xt = self.X.T 

  def betas(self):
    be = np.linalg.inv((self.Xt @ self.X)) @ (self.Xt @ self.Y)  #Betas
    self.matXB= self.X @ be                                      #resultado de X * betas
    self.mediaY = np.mean(self.Y)                                #Media Y
    self.U = (self.Y)-(self.matXB)                               #residuos
    self.SEC = np.sum(((self.matXB) - (self.mediaY) )**2)        #Suma explicada de Cudrados
    self.SRC = np.sum((self.U)**2)                               #Suma residual de cuadrados
    self.STC = self.SEC + self.SRC                               #Suma total de cuadrados
    self.R_cua = self.SEC/self.STC                               #R-cuadrado
    return be

  def grados_lib(self):
    gl = self.n - self.k                                         #Grados de libertad
    return gl

  def covarianza(self):
   cov = (self.SRC/(reg_lin.grados_lib(self))) * (np.linalg.inv((self.Xt @ self.X)))
   return cov

  def varianza(self):
   var = np.diag(reg_lin.covarianza(self))
   return var

  def errores_est(self):
    ee=np.sqrt(reg_lin.varianza(self))
    return ee

  def t_stat(self):
    ts = reg_lin.betas(self)/reg_lin.errores_est(self)
    return ts
  
  def conf_int_sup(self):
    sup = reg_lin.betas(self).T.squeeze()+1.96*reg_lin.errores_est(self)
    return sup

  def conf_int_inf(self):
    inf = reg_lin.betas(self).T.squeeze()-1.96*reg_lin.errores_est(self)
    return inf
```

```python
tabla1 = reg_lin (df['narr86'],df[["pcnv", "avgsen", "tottime", "inc86"]])
```

```python
tb=pd.DataFrame({'variables':['Constante','pcnv','avgsen','tottime','inc86'],
                 'betas': tabla1.betas(),
                 'errores est': tabla1.errores_est(),
                 't_stat': tabla1.t_stat(),
                 'limite superior':tabla1.conf_int_sup() ,
                 'limite inferior': tabla1.conf_int_inf()})
```
